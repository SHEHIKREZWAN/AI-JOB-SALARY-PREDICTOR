#   Outlier Detection & Treatment

# IQR Calculation

```python
Q1 = df['salary_usd'].quantile(0.25)
Q3 = df['salary_usd'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
```

###  Before treatment

```python
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

axes[0].boxplot(df['salary_usd'], vert=False, patch_artist=True,
                boxprops=dict(facecolor='#FFCDD2', color='red'),
                medianprops=dict(color='red'))
axes[0].set_title('salary_usd — Before Outlier Treatment', fontweight='bold')
axes[0].set_xlabel('Salary (USD)')

axes[1].boxplot(df['salary_usd'], vert=False, patch_artist=True,
                boxprops=dict(facecolor='#C8E6C9', color='green'),
                medianprops=dict(color='green'))
axes[1].set_title('salary_usd — After Outlier Treatment (Capping)', fontweight='bold')
axes[1].set_xlabel('Salary (USD)')
```

```python
plt.tight_layout()
plt.show()
print(f"\n Outliers capped at upper bound: ${upper_bound:,.0f}")
print(f" Dataset shape remains: {df.shape}")
```

# Frequency Encoding

```python
freq_cols = ['job_title', 'salary_currency', 'employment_type',
             'company_location', 'employee_residence', 'industry']
```

# ✅ Sprint 1 Addition: Save Preprocessing Pipeline
> We save all encoding maps + the fitted scaler so Sprint 4 can reproduce the SAME transformations on raw data without re-running Sprint 1.


```python
import joblib, os, json as _json

save_dir = '/Users/rezwanshehik/Downloads/'
os.makedirs(save_dir, exist_ok=True)

# 1. Save encoding maps (used in ordinal & frequency encoding)
encoding_maps = {
    'exp_map'  : {'EN': 0, 'MI': 1, 'SE': 2, 'EX': 3},
    'size_map' : {'S': 0, 'M': 1, 'L': 2},
    'edu_map'  : {'Associate': 0, 'Bachelor': 1, 'Master': 2, 'PhD': 3},
    'freq_cols': ['job_title', 'salary_currency', 'employment_type',
                  'company_location', 'employee_residence', 'industry'],
    'freq_maps': {col: X_train_final[col].to_dict()
                  if col in X_train_final.columns
                  else X_train[col].value_counts(normalize=True).to_dict()
                  for col in ['job_title', 'salary_currency', 'employment_type',
                               'company_location', 'employee_residence', 'industry']}
}

# Save freq maps (learnt from training data only)
freq_maps_save = {}
for col in ['job_title', 'salary_currency', 'employment_type',
            'company_location', 'employee_residence', 'industry']:
    freq_maps_save[col] = X_train[col].value_counts(normalize=True).to_dict() \
                          if col in X_train.columns else {}

encoding_maps['freq_maps'] = freq_maps_save

with open(save_dir + 'encoding_maps.json', 'w') as fp:
    _json.dump(encoding_maps, fp)
print('✅ encoding_maps.json saved')

# 2. Save fitted StandardScaler
joblib.dump(scaler, save_dir + 'scaler_sprint1.joblib')
print('✅ scaler_sprint1.joblib saved')

# 3. Save feature column order (critical for pipeline)
feature_cols = list(X_train.columns)
with open(save_dir + 'feature_columns.json', 'w') as fp:
    _json.dump(feature_cols, fp)
print(f'✅ feature_columns.json saved  ({len(feature_cols)} features)')
print(f'   Columns: {feature_cols}')
```

### Sprint 1 now saves 3 artefacts:
| File | Purpose |
|------|--------|
| `encoding_maps.json` | Ordinal & frequency encoding maps |
| `scaler_sprint1.joblib` | Fitted StandardScaler |
| `feature_columns.json` | Column order for pipeline input |
